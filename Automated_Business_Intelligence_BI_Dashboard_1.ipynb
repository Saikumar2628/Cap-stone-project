# 2. Combining all 9 datasets to CRM and ERP

```python
# Loading datasets
data_customers = pd.read_csv('raw_data/olist_customers_dataset.csv')
data_location = pd.read_csv('raw_data/olist_geolocation_dataset.csv')
data_reviews = pd.read_csv('raw_data/olist_order_reviews_dataset.csv')
data_items = pd.read_csv('raw_data/olist_order_items_dataset.csv')
data_payments = pd.read_csv('raw_data/olist_order_payments_dataset.csv')
data_orders = pd.read_csv('raw_data/olist_orders_dataset.csv')
data_products = pd.read_csv('raw_data/olist_products_dataset.csv')
data_sellers = pd.read_csv('raw_data/olist_sellers_dataset.csv')
data_product_cat = pd.read_csv('raw_data/product_category_name_translation.csv')

# Removing whitespace from column names
data_customers.columns = data_customers.columns.str.strip()
data_location.columns = data_location.columns.str.strip()
data_reviews.columns = data_reviews.columns.str.strip()
data_items.columns = data_items.columns.str.strip()
data_payments.columns = data_payments.columns.str.strip()
data_orders.columns = data_orders.columns.str.strip()
data_products.columns = data_products.columns.str.strip()
data_sellers.columns = data_sellers.columns.str.strip()
data_product_cat.columns = data_product_cat.columns.str.strip()

# Combine CRM datasets
# Merge customers with location
crm_df = pd.merge(data_customers, data_location, how='left', left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix')

# Merge CRM data with orders and reviews
# Orders data may not have 'order_id' in CRM, hence merge on 'customer_id' to get order IDs
crm_df = pd.merge(crm_df, data_orders[['order_id', 'customer_id']], how='left', on='customer_id')

# Merge CRM data with reviews
crm_df = pd.merge(crm_df, data_reviews[['order_id', 'review_score']], how='left', on='order_id')

# Combine ERP datasets
# Merge items with orders
erp_df = pd.merge(data_items, data_orders, how='left', on='order_id')

# Merge ERP data with payments
erp_df = pd.merge(erp_df, data_payments, how='left', on='order_id')

# Merge ERP data with products
erp_df = pd.merge(erp_df, data_products, how='left', on='product_id')

# Merge ERP data with sellers
erp_df = pd.merge(erp_df, data_sellers, how='left', on='seller_id')

# Merge ERP data with product categories
erp_df = pd.merge(erp_df, data_product_cat, how='left', on='product_category_name')

# Fill missing values in 'product_category_name' with 'product_category_name_english'
erp_df['product_category_name_english'].fillna('Unknown', inplace=True)
erp_df['product_category_name'].fillna(erp_df['product_category_name_english'], inplace=True)

# Drop the 'product_category_name_english' column if not needed
erp_df.drop(columns=['product_category_name_english'], inplace=True)

```

```python
# 1.Preparing Data for Anomaly Detection
#Feature Scaling:

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
erp_df_scaled = scaler.fit_transform(erp_df_cleaned[['price', 'freight_value', 'payment_value',  
                                                     'product_weight_g', 'product_length_cm',
                                                     'product_height_cm', 'product_width_cm']])
crm_df_scaled = scaler.fit_transform(crm_df_cleaned[['review_score']])
```

```python
erp_df_scaled, crm_df_scaled
```

```python
#2.Anomaly Detection Models

# a. Z-Score / Standard Deviation Method 


from sklearn.preprocessing import StandardScaler
import numpy as np

# Feature Scaling using StandardScaler (Z-Score)
scaler = StandardScaler()

# Scale ERP data
erp_features = ['price', 'freight_value', 'payment_value', 'product_weight_g', 
                'product_length_cm', 'product_height_cm', 'product_width_cm']
erp_df_scaled = scaler.fit_transform(erp_df_cleaned[erp_features])

# Scale CRM data
crm_features = ['review_score']
crm_df_scaled = scaler.fit_transform(crm_df_cleaned[crm_features])

# Z-Score based anomaly detection (thresholding method)
z_threshold = 3  # You can adjust this threshold as needed
erp_z_scores = np.abs(erp_df_scaled) > z_threshold
crm_z_scores = np.abs(crm_df_scaled) > z_threshold

# Adding the Z-Score anomaly detection results to the DataFrame
erp_df_cleaned['anomaly_z_score'] = np.any(erp_z_scores, axis=1).astype(int)
crm_df_cleaned['anomaly_z_score'] = crm_z_scores.astype(int)
```

```python
#b. Isolation Forest 

from sklearn.ensemble import IsolationForest

# Applying Isolation Forest on ERP scaled data
iso_forest_erp = IsolationForest(contamination=0.05, random_state=42)
erp_anomaly_pred = iso_forest_erp.fit_predict(erp_df_scaled)

# Applying Isolation Forest on CRM scaled data
iso_forest_crm = IsolationForest(contamination=0.05, random_state=42)
crm_anomaly_pred = iso_forest_crm.fit_predict(crm_df_scaled)

# Adding Isolation Forest anomaly detection results to the DataFrame
erp_df_cleaned['anomaly_isolation_forest'] = (erp_anomaly_pred == -1).astype(int)
crm_df_cleaned['anomaly_isolation_forest'] = (crm_anomaly_pred == -1).astype(int)
```

```python
from sklearn.cluster import DBSCAN
import pandas as pd
import numpy as np

# Assuming `erp_df_scaled` and `crm_df_scaled` are preprocessed NumPy arrays
# Assuming `erp_df_cleaned` and `crm_df_cleaned` are the original DataFrames corresponding to scaled data

# Define sample size
sample_size = 100  # Adjust based on your data size

# Generate random sample indices
erp_indices = np.random.choice(erp_df_scaled.shape[0], size=sample_size, replace=False)
crm_indices = np.random.choice(crm_df_scaled.shape[0], size=sample_size, replace=False)

# Create samples using the sampled indices
erp_sample = erp_df_scaled[erp_indices]
crm_sample = crm_df_scaled[crm_indices]

# Map sampled indices back to the original DataFrame's index
erp_original_indices = erp_df_cleaned.index[erp_indices]
crm_original_indices = crm_df_cleaned.index[crm_indices]

# Check and handle NaN values in the samples
if np.isnan(erp_sample).any():
    print("NaN values found in ERP sample. Please clean your data.")
else:
    # Apply DBSCAN on ERP scaled sample
    dbscan_erp = DBSCAN(eps=0.5, min_samples=5)  # Hyperparameters can be tuned
    erp_anomaly_dbscan = dbscan_erp.fit_predict(erp_sample)

    # Create anomaly column in the original DataFrame
    erp_df_cleaned['anomaly_dbscan'] = 0  # Initialize with 0 (non-anomalies)
    erp_df_cleaned.loc[erp_original_indices, 'anomaly_dbscan'] = (erp_anomaly_dbscan == -1).astype(int)  # Mark anomalies

if np.isnan(crm_sample).any():
    print("NaN values found in CRM sample. Please clean your data.")
else:
    # Apply DBSCAN on CRM scaled sample
    dbscan_crm = DBSCAN(eps=0.5, min_samples=5)  # Hyperparameters can be tuned
    crm_anomaly_dbscan = dbscan_crm.fit_predict(crm_sample)

    # Create anomaly column in the original DataFrame
    crm_df_cleaned['anomaly_dbscan'] = 0  # Initialize with 0 (non-anomalies)
    crm_df_cleaned.loc[crm_original_indices, 'anomaly_dbscan'] = (crm_anomaly_dbscan == -1).astype(int)  # Mark anomalies

# Display results
print("ERP DataFrame with Anomalies:")
print(erp_df_cleaned[['anomaly_dbscan']].head())

print("\nCRM DataFrame with Anomalies:")
print(crm_df_cleaned[['anomaly_dbscan']].head())
```
